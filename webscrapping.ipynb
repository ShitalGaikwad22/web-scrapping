```python
import requests
from bs4 import BeautifulSoup
```

```python
url= 'https://www.imdb.com/title/tt0386676/episodes'
```

```python
episodes = []
rating = []
```

```python
for season in range(1,10):
  r= requests.get(url, params={'season':season})
  soup = BeautifulSoup(r.text,'html5lib')
  listing = soup.find('div', class_='eplist')
  for epno, div in enumerate(listing.find_all('div',recursive=False)):
    eps = "{}.{}".format(season,epno+1)
    rel = div.find(class_='ipl-rating-star__rating') 
    ratings = float(rel.get_text(strip=True))
    print('Episode:',eps,'--Rating:',ratings)
    episodes.append(eps)
    rating.append(ratings)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
episodes = ['S' + e.split('.')[0] if int(e.split('.')[1]) == 1 else '' for e in episodes]
```

```python
episodes
```

```python
plt.figure(figsize=(15,6))
positions = [a*2 for a in range(len(rating))]
plt.bar(positions, ratings, align='center')
plt.xticks(positions, episodes)
plt.show()
```
